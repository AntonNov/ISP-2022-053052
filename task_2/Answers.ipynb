````python
start, stop, step = -10, 10, 3

#########################
#    -{function_name} object at ...
print(" range()")
list1 = list(range(start, stop, step))
print(list1)

print(" filter()")
print(list(filter(lambda x: x % 2 == 0, list1)))

print(" zip()")
#    ,        
list2 = list(range(7))
print(list(zip(list1, list2)))

print(" map()")
print(list(map(abs, list1)))
########################


print(" sorted()")
print(sorted(list1, reverse = True))
      
print(" any()")
print(any(el % 5 == 0 for el in list1))
      
print(" all()")
print(all(el % 5 == 0 for el in list1))
      
print(" sum()")
print(sum(list1))
      
print(" min()")
print(min(list1))
      
print(" max()")
print(max(list1))

print(" enumerate()  format")
for index, el in enumerate(list1):
    print(f"{index} {el}")
    
print(chr(97))
print(ord('a'))
````

## 6. Scopes (LEGB), global, nonlocal
L  Local.     ( / ),    (  def  lambda),        global.

E  Enclosing.            (,   def  lambda).

G  Global.    ,            global.

B  Build-in .   ,   print, input, open  .

````python
def counter(start=0):
    def step():
        nonlocal start
        start += 1
        return start
    
    return step


print(counter(6), counter())
````

````python
def my_shiny_new_decorator(function_to_decorate):
     #    -"".     ,
     #        .
     def the_wrapper_around_the_original_function():
         print(",    ")
         function_to_decorate() #  
         print(",    \n")
     #   
     return the_wrapper_around_the_original_function

 #  ,     ,      .
def simple_function():
     print("Hello")

simple_function()
print()

""" ,    ,    ,     ,
       ,   ,   ,
    :"""
simple_function = my_shiny_new_decorator(simple_function)
simple_function()


"""   
       @."""
@my_shiny_new_decorator
def another_simple_function():
     print("Hello world")

another_simple_function()
````

````python
"""   ,    .
      -      ."""

from random import randint


def my_decorator(func):
    def the_wrapper(my_list):
        func([elem for elem in my_list if elem % 2])  #  

    return the_wrapper  #   


@my_decorator
def my_func(my_list):
    print(my_list)


my_list = [randint(-100, 100) for _ in range(20)]
print(my_list)
my_func(my_list)
````

````python
"""  ,        ."""


def my_decorator(my_func):
    def wrapper(*args):
        my_func(args[::-1])  #   

    return wrapper


@my_decorator
def print_args(*args):
    [print(*arg) for arg in args]


args = ("first", 1, [4], 6, 8, {1: True}, "a", "last")
print(*args)
print_args(*args)
````

##   

+ **@classmethod** -    ..  .
+ **@staticmethod** -      .
+ **@classmethod**       ,     ,     .     **@staticmethod** ,        ,     .
+ **@staticmethod**     ,        ,     .     ,     **self**,      .

````python
class SomeClass:
    def __init__(self, value):
        self.value = value
    @classmethod
    def print_name(cls):
        #self,     .
        #cls,    ."""
        print(cls.__name__)

class ChildClass(SomeClass):
    pass

SomeClass.print_name()
ChildClass.print_name()
````

````python
class SomeOtherClass:
    @staticmethod
    def print_hello():
        print("Hello!")

SomeOtherClass.print_hello()
````

````python
def decorator_maker():
    print(" .        ")
    
    def my_decorator(func):
        print(".       ")
            
        def wrapped():
            print("")
            return func()
        
        print("  ")
        return wrapped
    
    print(" ")
    return my_decorator


@decorator_maker()
def simple_function():
    print("  ")

print("\n\n")
simple_function()
````
